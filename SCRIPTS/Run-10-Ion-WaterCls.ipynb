```python
import os
import ion_gen as ig
import glob
```

```python
psi4Exe = "/home/kchung25/anaconda3/envs/psi4_19/bin/psi4"
qchemExe = "/home/kchung25/qchem/5.4.0/bin/qchem"

cwd = os.getcwd()
```

```python
# MP2

mp2Dir = f"{cwd}/../MP2"
os.chdir(mp2Dir)

for key in ig.ionClsWater.keys():
    index = ig.index[key]
    frames = ig.allTwoBody[key]
    for frame in frames:
        fileName = f"{index}_{key}_{str(frame).zfill(3)}_mp2.in"
        os.system(f"{psi4Exe} -n 6 {fileName}")
```

```python
# SAPT2+

saptDir = f"{cwd}/../SAPT2+"
os.chdir(saptDir)

for key in ig.ionClsWater.keys():
    index = ig.index[key]
    frames = ig.allTwoBody[key]
    if "K" not in key:
        continue
    for frame in frames:
        fileName = f"{index}_{key}_{str(frame).zfill(3)}_sapt.in"
        os.system(f"{psi4Exe} -n 6 {fileName}")
```

```python
# ALMO

almoDir = f"{cwd}/../ALMO"
os.chdir(almoDir)

for key in ig.ionClsWater.keys():
    index = ig.index[key]
    frames = ig.allTwoBody[key]
    for frame in frames:
        fileNameIn = f"{index}_{key}_{str(frame).zfill(3)}_almo.in"
        fileNameOut = f"{index}_{key}_{str(frame).zfill(3)}_almo.out"
        os.system(f"{qchemExe} -np 6 -nt 6 {fileNameIn} {fileNameOut}")
```
