```python
import os
import ion_gen as ig
```

```python
cwd = os.getcwd()
```

```python
def splitDimer(dimer):
    if dimer == "LiF":
        ion1 = "Li"
        ion2 = "F"
    elif dimer == "LiCl":
        ion1 = "Li"
        ion2 = "Cl"
    elif dimer == "LiBr":
        ion1 = "Li"
        ion2 = "Br"
    elif dimer == "NaF":
        ion1 = "Na"
        ion2 = "F"
    elif dimer == "NaCl":
        ion1 = "Na"
        ion2 = "Cl"
    elif dimer == "NaBr":
        ion1 = "Na"
        ion2 = "Br"
    elif dimer == "KF":
        ion1 = "K"
        ion2 = "F"
    elif dimer == "KCl":
        ion1 = "K"
        ion2 = "Cl"
    elif dimer == "KBr":
        ion1 = "K"
        ion2 = "Br"
    return ion1,ion2
```

```python
# Tinker

xyzDir = f"{cwd}/../XYZ"
os.chdir(xyzDir)

for key,item in ig.ion2Ion.items():
    split = key.split("-")
    dimer = split[0]
    ion1,ion2 = splitDimer(dimer)
    ion3 = split[1]
    at1 = ig.atomType[ion1]
    at2 = ig.atomType[ion2]
    at3 = ig.atomType[ion3]
    index = ig.index[key]
    distances = ig.allTwoBody[key]
    if ion1 in ["Li","Na","K"] and ion3 in ["Li","Na","K"]:
        ion = ion1
        at = at1
        ion1 = ion2
        at1 = at2
        ion2 = ion
        at2 = at
    for dist in distances:
        fileName = f"{index}_{key}_{dist:3.2f}.xyz"
        f = open(fileName,"w")
        f.write(ig.ion2IonXyz.format(ion1,at1,ion2,item,at2,ion3,item*dist,at3))
        f.close()
```

```python
# MP2

mp2Dir = f"{cwd}/../MP2"
os.chdir(mp2Dir)

for key,item in ig.ion2Ion.items():
    split = key.split("-")
    dimer = split[0]
    ion1,ion2 = splitDimer(dimer)
    ion3 = split[1]
    ac1 = ig.atomCharge[ion1]
    ac2 = ig.atomCharge[ion2]
    ac3 = ig.atomCharge[ion3]
    index = ig.index[key]
    distances = ig.allTwoBody[key]
    if ion1 in ["Li","Na","K"] and ion3 in ["Li","Na","K"]:
        ion = ion1
        at = at1
        ion1 = ion2
        at1 = at2
        ion2 = ion
        at2 = at
    for dist in distances:
        fileName = f"{index}_{key}_{dist:3.2f}_mp2.in"
        f = open(fileName,"w")
        f.write(ig.psi4Header.format(ig.memory))
        f.write(ig.psi4Molecule.format(ig.ion2IonPsi4.format(ac1+ac2,ion1,ion2,item,ac3,ion3,item*dist)))
        f.write(ig.psi4Set.format(ig.mp2SetPsi4))
        f.write(ig.psi4Final.format(ig.mp2FinalPsi4))
        f.close()
```
