```python
import os
import ion_gen as ig
```

```python
psi4Exe = "/home/kchung25/anaconda3/envs/psi4_19/bin/psi4"
qchemExe = "/home/kchung25/qchem/5.4.0/bin/qchem"

cwd = os.getcwd()
```

```python
# MP2

mp2Dir = f"{cwd}/../MP2"
os.chdir(mp2Dir)

for key in ig.ionDegWater:
    index = ig.index[key]
    angles = ig.allTwoBody[key]
    for ang in angles:
        sign = "+"
        if ang < 0:
            sign = "-"
        fileName = f"{index}_{key}_{sign}{str(abs(ang)).zfill(2)}_mp2.in"
        os.system(f"{psi4Exe} -n 6 {fileName}")
```

```python
# SAPT2+

saptDir = f"{cwd}/../SAPT2+"
os.chdir(saptDir)

for key in ig.ionDegWater:
    index = ig.index[key]
    angles = ig.allTwoBody[key]
    for ang in angles:
        sign = "+"
        if ang < 0:
            sign = "-"
        fileName = f"{index}_{key}_{sign}{str(abs(ang)).zfill(2)}_sapt.in"
        os.system(f"{psi4Exe} -n 6 {fileName}")
```

```python
# ALMO

almoDir = f"{cwd}/../ALMO"
os.chdir(almoDir)

for key in ig.ionDegWater:
    index = ig.index[key]
    angles = ig.allTwoBody[key]
    for ang in angles:
        sign = "+"
        if ang < 0:
            sign = "-"
        fileNameIn = f"{index}_{key}_{sign}{str(abs(ang)).zfill(2)}_almo.in"
        fileNameOut = f"{index}_{key}_{sign}{str(abs(ang)).zfill(2)}_almo.out"
        os.system(f"{qchemExe} -np 6 -nt 6 {fileNameIn} {fileNameOut}")
```
