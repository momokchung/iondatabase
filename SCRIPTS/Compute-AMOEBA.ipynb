```python
baseDir = f"{cwd}/.."
os.chdir(baseDir)

amoebaHeader = "Filename                           Electrostatics             VdW    Polarization          AMOEBA\n"
amoebaTemp = "{:33}{:16.6f}{:16.6f}{:16.6f}{:16.6f}\n"

all_fn = [
    fn01,
    fn02,
    fn03,
    fn04,
    fn05,
    fn06,
    fn07,
    fn08,
    fn09,
    fn10,
]

all_compute = [
    compute01,
    compute02,
    compute03,
    compute04,
    compute05,
    compute06,
    compute07,
    compute08,
    compute09,
    compute10,
]

lines = []
for i in range(10):
    fni = all_fn[i]
    computei = all_compute[i]
    index = i+1
    for j in range(len(fni)):
        fn = fni[j]
        compute = computei[j]
        eint = compute.int
        epot = compute.pot
        evdw = compute.vdw
        eele = compute.ele
        epol = compute.pol
        if index in [2,5,8]:
            e = eint
        else:
            e = epot
        lines.append(amoebaTemp.format(fn, eele, evdw, epol, e))

def extract_number(s):
    parts = s.split("_")
    return int(parts[0])

# Sort the list based on the number
lines = sorted(lines, key=extract_number)

fw = open("ions.AMOEBA.dat","w")
fw.write(amoebaHeader)
for line in lines:
    fw.write(line)
fw.close()
```
