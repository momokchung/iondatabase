```python
import ion_gen as ig
import os
```

```python
twoBodyTxt = ""
for i,k in ig.allTwoBody.items():
    if len(k) == 1:
        twoBodyTxt += f"{ig.index[i]}_{i}\n"
    else:
        if "Deg" in i:
            for dist in k:
                sign = "+"
                if dist < 0:
                    sign = "-"
                twoBodyTxt +=f"{ig.index[i]}_{i}_{sign}{str(abs(dist)).zfill(2)}\n"
        else:
            for dist in k:
                twoBodyTxt += f"{ig.index[i]}_{i}_{dist:3.2f}\n"
```

```python
threeBodyTxt = ""
for i,k in ig.allThreeBody.items():
    if len(k) == 1:
        threeBodyTxt += f"{ig.index[i]}_{i}\n"
    else:
        if "Deg" in i:
            for dist in k:
                sign = "+"
                if dist < 0:
                    sign = "-"
                threeBodyTxt +=f"{ig.index[i]}_{i}_{sign}{str(abs(dist)).zfill(2)}\n"
        else:
            for dist in k:
                threeBodyTxt +=f"{ig.index[i]}_{i}_{dist:3.2f}\n"
```
