```python
import os
import ion_gen as ig
```

```python
ion = "Br"

cwd = os.getcwd()
```

```python
indexDict = {
    "Li": "01",
    "Na": "02",
    "K":  "03",
    "F":  "04",
    "Cl": "05",
    "Br": "06",
}

innerDict = {
    "Li": 1.895,
    "Na": 2.365,
    "K":  2.765,
    "F":  2.825,
    "Cl": 3.195,
    "Br": 3.325,
}

outerDict = {
    "Li": 2.805,
    "Na": 3.285,
    "K":  3.595,
    "F":  3.355,
    "Cl": 3.865,
    "Br": 4.405,
}

tinkerDict = {
    "Li": "Li+",
    "Na": "Na+",
    "K":  "K+" ,
    "F":  "F-" ,
    "Cl": "Cl-",
    "Br": "Br-",
}

atDict = {
    "Li": 351,
    "Na": 352,
    "K":  353,
    "F":  362,
    "Cl": 363,
    "Br": 364,
}

fileLength = 2703
frames = 1000
```

```python
def image(xr, yr, zr, xcell, ycell, zcell):
    xcell2 = xcell / 2
    ycell2 = ycell / 2
    zcell2 = zcell / 2
    while abs(xr) > xcell2:
        xr -= abs(xcell) if xr > 0 else -abs(xcell)
    while abs(yr) > ycell2:
        yr -= abs(ycell) if yr > 0 else -abs(ycell)
    while abs(zr) > zcell2:
        zr -= abs(zcell) if zr > 0 else -abs(zcell)
    return xr, yr, zr

ionTemp = "{:>6d}  {:3}{:12.6f}{:12.6f}{:12.6f}{:>6d}\n"
OTemp = "{:>6d}  {:3}{:12.6f}{:12.6f}{:12.6f}{:>6d}{:>6d}{:>6d}\n"
HTemp = "{:>6d}  {:3}{:12.6f}{:12.6f}{:12.6f}{:>6d}{:>6d}\n"
```

```python
index = indexDict[ion]

cut = innerDict[ion]+0.3
cut2 = cut*cut

tinker = tinkerDict[ion]
at = atDict[ion]

clsIndex = ig.index[f"{ion}-Water_Cls"]

fn = f"{cwd}/../SIM/{index}_{ion}/{ion}water.arc"

f = open(fn, "r")
lines = f.readlines()
f.close()

saveIndex = 0
for frame in range(frames-100,frames):
    xyz = lines[frame*fileLength:(frame+1)*fileLength]
    split = xyz[1].split()
    xcell = float(split[0])
    ycell = float(split[1])
    zcell = float(split[2])
    split = xyz[2].split()
    xi = float(split[2])
    yi = float(split[3])
    zi = float(split[4])
    counter = 1
    txt = ionTemp.format(counter, tinker, 0, 0, 0, at)
    counter += 1
    for i in range(3, len(xyz), 3):
        split = xyz[i].split()
        xk = float(split[2])
        yk = float(split[3])
        zk = float(split[4])
        xr = xi - xk
        yr = yi - yk
        zr = zi - zk
        xr,yr,zr = image(xr,yr,zr,xcell,ycell,zcell)
        r2 = xr*xr + yr*yr + zr*zr
        if (r2 < cut2):
            txt += OTemp.format(counter, "O", xr, yr, zr, 349, counter+1, counter+2)
            split = xyz[i+1].split()
            xk = float(split[2])
            yk = float(split[3])
            zk = float(split[4])
            xr = xi - xk
            yr = yi - yk
            zr = zi - zk
            xr,yr,zr = image(xr,yr,zr,xcell,ycell,zcell)
            txt += HTemp.format(counter+1, "H", xr, yr, zr, 350, counter)
            split = xyz[i+2].split()
            xk = float(split[2])
            yk = float(split[3])
            zk = float(split[4])
            xr = xi - xk
            yr = yi - yk
            zr = zi - zk
            xr,yr,zr = image(xr,yr,zr,xcell,ycell,zcell)
            txt += HTemp.format(counter+2, "H", xr, yr, zr, 350, counter)
            counter += 3
    txt = f"{counter-1:>6d}\n" + txt
    print((counter-2)/3)
    fn = f"{cwd}/../SIM/{index}_{ion}/CLUSTER/{clsIndex}_{ion}-Water_Cls_{str(saveIndex).zfill(3)}.xyz"
    f = open(fn, "w")
    f.write(txt)
    f.close()
    saveIndex += 1
```
