```python
import os
import ion_gen as ig
import structure as st
```

```python
cwd = os.getcwd()
```

```python
# MP2

mp2Dir = f"{cwd}/../MP2"
os.chdir(mp2Dir)

for key,value in ig.ionClsWater.items():
    split = key.split("-")
    ion = split[0]
    ac = ig.atomCharge[ion]
    index = ig.index[key]
    frames = ig.allTwoBody[key]
    for frame in frames:
        fn = f"{cwd}/../SIM/{value}_{ion}/CLUSTER/{index}_{key}_{str(frame).zfill(3)}.xyz"
        f = open(fn, "r")
        lines = f.readlines()
        f.close()
        split = lines[1].split()
        x = float(split[2])
        y = float(split[3])
        z = float(split[4])
        ionTxt = ig.atomCLSTemp.format(ion,x,y,z)
        ionGhTxt = ig.atomCLSTemp.format(f"Gh({ion})",x,y,z)
        watersTxt = ""
        watersGhTxt = ""
        split = lines[2].split()
        at = split[1]
        x = float(split[2])
        y = float(split[3])
        z = float(split[4])
        watersTxt += ig.atomCLSTemp.format(at,x,y,z)
        watersGhTxt += ig.atomCLSTemp.format(f"Gh({at})",x,y,z)
        split = lines[3].split()
        at = split[1]
        x = float(split[2])
        y = float(split[3])
        z = float(split[4])
        watersTxt += ig.atomCLSTemp.format(at,x,y,z)
        watersGhTxt += ig.atomCLSTemp.format(f"Gh({at})",x,y,z)
        split = lines[4].split()
        at = split[1]
        x = float(split[2])
        y = float(split[3])
        z = float(split[4])
        watersTxt += ig.atomCLSTemp.format(at,x,y,z)
        watersGhTxt += ig.atomCLSTemp.format(f"Gh({at})",x,y,z)
        watersLst = []
        watersGhLst = []
        for line in lines[5:]:
            split = line.split()
            at = split[1]
            if at == "O":
                watersLst.append(watersTxt)
                watersGhLst.append(watersGhTxt)
                watersTxt = ""
                watersGhTxt = ""
            x = float(split[2])
            y = float(split[3])
            z = float(split[4])
            watersTxt += ig.atomCLSTemp.format(at,x,y,z)
            watersGhTxt += ig.atomCLSTemp.format(f"Gh({at})",x,y,z)
        watersLst.append(watersTxt)
        watersGhLst.append(watersGhTxt)
        nWater = len(watersLst)
        watersTxt = ""
        for text in watersLst:
            watersTxt += text
        watersGhTxt = ""
        for text in watersGhLst:
            watersGhTxt += text
        txt = ig.tempCLSPsi4.format(ac,ionTxt+watersTxt)
        fileName = f"{index}_{key}_CLS_{str(frame).zfill(3)}_mp2.in"
        f = open(fileName,"w")
        f.write(ig.psi4Header.format(ig.memory))
        f.write(ig.psi4Molecule.format(txt))
        f.write(ig.psi4Set.format(ig.mp2SetPsi4))
        f.write(ig.psi4Final.format(ig.mp2FinalNoCPPsi4))
        f.close()
        txt = ig.tempCLSPsi4.format(ac,ionTxt+watersGhTxt)
        fileName = f"{index}_{key}_CLS_{str(frame).zfill(3)}_00_mp2.in"
        f = open(fileName,"w")
        f.write(ig.psi4Header.format(ig.memory))
        f.write(ig.psi4Molecule.format(txt))
        f.write(ig.psi4Set.format(ig.mp2SetPsi4))
        f.write(ig.psi4Final.format(ig.mp2FinalNoCPPsi4))
        f.close()
        for i in range(nWater):
            tmpTxt = ""
            for j in range(nWater):
                if i == j:
                    tmpTxt += watersLst[j]
                else:
                    tmpTxt += watersGhLst[j]
            txt = ig.tempCLSPsi4.format(0,ionGhTxt+tmpTxt)
            fileName = f"{index}_{key}_CLS_{str(frame).zfill(3)}_{str(i+1).zfill(2)}_mp2.in"
            f = open(fileName,"w")
            f.write(ig.psi4Header.format(ig.memory))
            f.write(ig.psi4Molecule.format(txt))
            f.write(ig.psi4Set.format(ig.mp2SetPsi4))
            f.write(ig.psi4Final.format(ig.mp2FinalNoCPPsi4))
            f.close()
        
```
