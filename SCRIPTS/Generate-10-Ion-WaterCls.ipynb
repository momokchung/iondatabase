```python
import os
import ion_gen as ig
import structure as st
```

```python
cwd = os.getcwd()
```

```python
# Tinker

xyzDir = f"{cwd}/../XYZ"
os.chdir(xyzDir)

for key,value in ig.ionClsWater.items():
    split = key.split("-")
    ion = split[0]
    index = ig.index[key]
    frames = ig.allTwoBody[key]
    for frame in frames:
        os.system(f"cp {cwd}/../SIM/{value}_{ion}/CLUSTER/{index}_{key}_{str(frame).zfill(3)}.xyz .")
```

```python
# MP2

mp2Dir = f"{cwd}/../MP2"
os.chdir(mp2Dir)

for key,value in ig.ionClsWater.items():
    split = key.split("-")
    ion = split[0]
    ac = ig.atomCharge[ion]
    index = ig.index[key]
    frames = ig.allTwoBody[key]
    for frame in frames:
        ionTxt = ""
        watersTxt = ""
        fn = f"{cwd}/../SIM/{value}_{ion}/CLUSTER/{index}_{key}_{str(frame).zfill(3)}.xyz"
        f = open(fn, "r")
        lines = f.readlines()
        f.close()
        split = lines[1].split()
        x = float(split[2])
        y = float(split[3])
        z = float(split[4])
        ionTxt += ig.atomTemp.format(ion,x,y,z)
        for line in lines[2:]:
            split = line.split()
            at = split[1]
            x = float(split[2])
            y = float(split[3])
            z = float(split[4])
            watersTxt += ig.atomTemp.format(at,x,y,z)
        txt = ig.tempPsi4.format(ac,ionTxt,0,watersTxt)
        fileName = f"{index}_{key}_{str(frame).zfill(3)}_mp2.in"
        f = open(fileName,"w")
        f.write(ig.psi4Header.format(ig.memory))
        f.write(ig.psi4Molecule.format(txt))
        f.write(ig.psi4Set.format(ig.mp2SetPsi4))
        f.write(ig.psi4Final.format(ig.mp2FinalPsi4))
        f.close()
```
