```python
import os
import ion_gen as ig
import compute as cm
```

```python
cwd = os.getcwd()
xyzDir = f"{cwd}/../XYZ/AMOEBA"
os.chdir(xyzDir)

exe = "~/tinker/source/analyze.x"
tkey = "~/iondatabase/PARAMS/amoebabio18.key"
```

```python
# Compute-03-CLS

fn03 = []
compute03 = []

for key,item in ig.ion2WaterEquil.items():
    index = ig.index[key]
    distances = ig.allTwoBody[key]
    for dist in distances:
        fn = f"{index}_{key}_{dist:3.2f}"
        xyz = f"{fn}.xyz"
        energy = cm.analyze(exe, tkey, xyz)
        fn03.append(fn)
        compute03.append(energy)
```

```python
# Compute-06-CLS

fn06 = []
compute06 = []

for key,item in ig.ion2Ion.items():
    index = ig.index[key]
    distances = ig.allTwoBody[key]
    for dist in distances:
        fn = f"{index}_{key}_{dist:3.2f}"
        xyz = f"{fn}.xyz"
        energy = cm.analyze(exe, tkey, xyz)
        fn06.append(fn)
        compute06.append(energy)
```

```python
# Compute-09-CLS

fn09 = []
compute09 = []

for key in ig.ion2WaterNeq:
    index = ig.index[key]
    distances = ig.allTwoBody[key]
    for dist in distances:
        fn = f"{index}_{key}_{dist:3.2f}"
        xyz = f"{fn}.xyz"
        energy = cm.analyze(exe, tkey, xyz)
        fn09.append(fn)
        compute09.append(energy)
```

```python
# Compute-10

fn10 = []
compute10 = []

for key,value in ig.ionClsWater.items():
    index = ig.index[key]
    frames = ig.allTwoBody[key]
    for frame in frames:
        fn = f"{index}_{key}_{str(frame).zfill(3)}"
        xyz = f"{fn}.xyz"
        energy = cm.analyze(exe, tkey, xyz)
        fn10.append(fn)
        compute10.append(energy)
```

```python
baseDir = f"{cwd}/.."
os.chdir(baseDir)

amoebaHeader = "Filename                                   AMOEBA\n"
amoebaTemp = "{:33}{:16.6f}\n"

fw = open("ions.AMOEBACLS.dat","w")
fw.write(amoebaHeader)

all_fn = [
    fn03,
    fn06,
    fn09,
    fn10,
]

all_compute = [
    compute03,
    compute06,
    compute09,
    compute10,
]

for i in range(4):
    fni = all_fn[i]
    computei = all_compute[i]
    for j in range(len(fni)):
        fn = fni[j]
        compute = computei[j]
        eint = compute.int
        epot = compute.pot
        evdw = compute.vdw
        eele = compute.ele
        epol = compute.pol
        # 06
        if i == 1:
            e = epot
        else:
            e = eint
        fw.write(amoebaTemp.format(fn, e))

fw.close()
```
