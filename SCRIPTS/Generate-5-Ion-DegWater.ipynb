```python
cwd = os.getcwd()

def rotate(vector, ax, deg):
    cos = np.cos(deg * np.pi/180)
    sin = np.sin(deg * np.pi/180)
    
    if ax == "x":
        mat = [
            [ 1.0, 0.0, 0.0],
            [ 0.0, cos,-sin],
            [ 0.0, sin, cos],
        ]
    elif ax == "y":
        mat = [
            [ cos, 0.0, sin],
            [ 0.0, 1.0, 0.0],
            [-sin, 0.0, cos],
        ]
    elif ax == "z":
        mat = [
            [ cos,-sin, 0.0],
            [ sin, cos, 0.0],
            [ 0.0, 0.0, 1.0],
        ]
    for vec in vector:
        v0 = vec[0]
        v1 = vec[1]
        v2 = vec[2]
        vec[0] = mat[0][0] * v0 + mat[0][1] * v1 + mat[0][2] * v2
        vec[1] = mat[1][0] * v0 + mat[1][1] * v1 + mat[1][2] * v2
        vec[2] = mat[2][0] * v0 + mat[2][1] * v1 + mat[2][2] * v2
    return
```
